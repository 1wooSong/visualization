```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size']=15
matplotlib.rcParams['axes.unicode_minus'] =False #마이너스 깨짐 현상 해결
df=pd.read_excel('score.xlsx')
df
```

```python
df['학년'] =[3,3,1,2,3,2,2,1]
df
```

```python
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
import numpy as np
sizes=np.random.rand(8)*1000
sizes
```

```python
plt.scatter(df['영어'],df['수학'],s=sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
sizes=df['학년']*200 # 1학년 500 2학년 1000 3 학년 1500
plt.scatter(df['영어'],df['수학'],s=sizes)
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis') #학년별로 색깔 다르게  #matplotlib cmap 사용하면 돼 
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
plt.figure(figsize=(7,7))
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis') #학년별로 색깔 다르게  #matplotlib cmap 사용하면 돼 
plt.xlabel('영어점수')
plt.ylabel('수학점수')
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.5,orientation='horizontal')
```
