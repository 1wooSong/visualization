```python
import pandas as pd 
df= pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df.dropna(inplace=True)# nan 포함한 데이터 삭제
```

```python
df
```

```python
df.dropna(axis='index',how='any')   #nan 이 하나라도 있는 row 삭제 
#axis  = index or columns
#how = any or all  any 면 ㄴ하나라도 nan 이 있으면 지우는거고 all 은 전체 데이터가 nan  일때 지우는거 
```
