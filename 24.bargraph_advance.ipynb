```python
labels= ['강백호','서태웅','정대만']
values=[190,186,174]

plt.barh(labels,values)
plt.xlim(170,190)
```

```python
bar=plt.bar(labels,values)
bar[0].set_hatch('/')#///
bar[1].set_hatch('x')
bar[2].set_hatch('..') #구글에서 matplotlib hatch 
```

```python
bar=plt.bar(labels,values)
plt.ylim(170,195)
for idx,rect in enumerate(bar):
    plt.text(idx,rect.get_height()+0.5, values[idx],ha='center',color='b')
```
