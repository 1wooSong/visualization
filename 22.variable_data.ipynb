```python
days=[1,2,3]
az=[2,4,8]
pfizer=[5,1,3]
moderna=[1,2,9]

plt.plot(days,az)
plt.plot(days,pfizer)
plt.plot(days,moderna)
```

```python
plt.plot(days,az,label='az')
plt.plot(days,pfizer,label='pf',marker='o',ls=':')
plt.plot(days,moderna,label='md',marker='v',ls='-.')
plt.legend()
```
